```python
from scipy import stats
import numpy as np

# Sample data for two groups
group1 = [85, 90, 78, 92, 88, 75, 95, 89]
group2 = [80, 83, 85, 87, 82, 79, 84, 86]

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Output the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Set significance level
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis. The means are significantly different.")
else:
    print("We fail to reject the null hypothesis. The means are not significantly different.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Sample data
data = [85, 88, 90, 92, 75, 85, 87, 89, 91, 83]

# Hypothesized population mean
population_mean = 80

# Calculate t-statistic and p-value
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Plot histogram of the sample data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=5, alpha=0.7, color='skyblue', edgecolor='black', label='Sample Data')

# Add vertical lines for sample mean and population mean
plt.axvline(sample_mean, color='blue', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(population_mean, color='red', linestyle='solid', linewidth=2, label=f'Population Mean: {population_mean:.2f}')

# Add confidence interval (95%) for the sample mean
confidence_interval = stats.t.interval(0.95, len(data)-1, loc=sample_mean, scale=sample_std/np.sqrt(len(data)))
plt.axvspan(confidence_interval[0], confidence_interval[1], alpha=0.2, color='blue', label='95% Confidence Interval')

# Title and labels
plt.title('One-Sample t-Test: Sample Data vs Population Mean', fontsize=15)
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()

# Show the plot
plt.show()
```

# paired t test

```python
before = np.random.normal(60,10,30)
after = before + np.random.normal(2,5,30)
t_statistic,p_value = stats.ttest_rel(before,after)
print(f"t-statistic:{t_statistic},p-value:{p_value}")

alpha = 0.05
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```

# z test

```python
from statsmodels.stats.weightstats import *
data = np.random.normal(100,15,50)

z_statistic, p_value = ztest(data, value=110)
print(f'z-statistic:{z_statistic},p-value:{p_value}')

alpha = 0.05
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
    
```

# anova 
anova tests weather the means of three or more independent groups are significantly different 

```python
group1 = np.random.normal(50,10,30)
group2 = np.random.normal(55,10,30)
group3 = np.random.normal(60,10,30)

f_statistic,p_value = stats.f_oneway(group1,group2,group3)
print(f"f-statistic:{f_statistic},p-value:{p_value}")

alpha = 0.05
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis ")
```

# chi square test 
a chi square test assess whether observed frequences in categorical data  differ from expected frequencies

```python
observed = np.array([20,30,25,25])
expected = np.array([25,25,25,25])
chi_statistic,p_value = stats.chisquare(f_obs=observed ,f_exp=expected)
print(f"chi-square statistic:{chi_statistic},p_value:{p_value}")

alpha = 0.05
if p_value < alpha:
    print("we reject the null hypothesis")
else:  
    print("we accept the null hypotheis")
```
