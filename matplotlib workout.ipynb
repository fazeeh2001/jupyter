```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Using Pyplot API
plt.plot(x, y_sin, label='Sine Wave', color='blue')   # Plot sine wave
plt.plot(x, y_cos, label='Cosine Wave', color='orange')  # Plot cosine wave

# Adding title and labels
plt.title("Sine and Cosine Waves")
plt.xlabel("X values (radians)")
plt.ylabel("Y values")

# Adding a legend
plt.legend()

# Show the plot
plt.grid()  # Optional: Add a grid for better visualization
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Using Object-Oriented API
fig, ax = plt.subplots()  # Create a figure and an axes object

# Plot sine and cosine waves
ax.plot(x, y_sin, label='Sine Wave', color='blue')   # Plot sine wave
ax.plot(x, y_cos, label='Cosine Wave', color='orange')  # Plot cosine wave

# Adding title and labels using Axes object
ax.set_title("Sine and Cosine Waves")
ax.set_xlabel("X values (radians)")
ax.set_ylabel("Y values")

# Adding a legend
ax.legend()

# Show the plot
ax.grid()  # Optional: Add a grid for better visualization
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn
df_sales = pd.DataFrame({
    "Quarter":["Q1 2022","Q2 2022","Q3 2022","Q4 2022","Q5 2022"],
    "Fridge":[12.5,11.0,13,16,15],
    "Dishwasher":[8,9,11,9.7,10.7],
    "washing machine":[9.2,8.5,7.5,7.6,6.8]
})
df_sales
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'])
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='green',label='fridge')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='yellow',label='Dishwasher')
plt.plot(df_sales['Quarter'],df_sales['washing machine'],color='blue',label='washing machine')
plt.title('product sales')
plt.ylabel('revenue')
plt.xlabel('Quarter')
plt.legend()
plt.show()
```

```python
total=df_sales[['Fridge','Dishwasher','washing machine']].sum()
total
```

```python
total.index
```

```python
plt.pie(total,labels=total.index,autopct='%1.3f%%',explode=(0.1,0.1,0),shadow=True,startangle=140)
plt.show()
```
