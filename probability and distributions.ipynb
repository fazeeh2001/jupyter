```python
import random

# Number of coin flips
n_flips = 50

# Initialize counters for heads and tails
count_heads = 0
count_tails = 0

# Simulate coin flips
for _ in range(n_flips):
    outcome = random.choice(['heads', 'tails'])
    
    # Count the outcomes
    if outcome == 'heads':
        count_heads += 1
    else:
        count_tails += 1

# Calculate probabilities
prob_heads = count_heads / n_flips
prob_tails = count_tails / n_flips

# Print the results
print("Number of Heads:", count_heads)
print("Number of Tails:", count_tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```

```python
from scipy.stats import bernoulli

# Probability of success
p = 0.7
# Create a Bernoulli random variable
bernoulli_dist = bernoulli(p)
# Calculate the probability of survival (success, k=1)
prob_surv = bernoulli_dist.pmf(1)
# Calculate the probability of non-survival (failure, k=0)
prob_not_surv = bernoulli_dist.pmf(0)

print(f"Probability of Survival: {prob_surv}")      # Output: 0.7
print(f"Probability of Non-Survival: {prob_not_surv}")  # Output: 0.3
```

# binomial distribution

```python
n = 10
p = 0.3
binomial_rv = stats.binom(n,p)
sample = binomial_rv.rvs(10)
sample
```

```python
x = np.arange(0,n)
pmf = binomial_rv.pmf(x)
pmf
```

```python
plt.bar(x,pmf)
plt.title('binomial distribution')
plt.xlabel('number of success')
plt.ylabel('probability')
plt.show()
```

# for normal distribution

```python
mu = 0
sigma = 1
normal_rv = stats.norm(mu ,sigma)
sample = normal_rv.rvs(size=1000)
sample
```

```python
#pdf
x = np.linspace(-3,3,100)
#print(x)
pdf = normal_rv.pdf(x)
plt.plot(x, pdf)
plt.title('normal distribution PDF')
plt.xlabel('value')
plt.ylabel('density')
plt.show()

```
