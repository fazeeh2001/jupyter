```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [5, 10, 7, 8]  # First subgroup
values2 = [3, 5, 2, 4]   # Second subgroup
values3 = [2, 3, 5, 6]   # Third subgroup

# Bar width
bar_width = 0.4

# Create a stacked bar chart
plt.bar(categories, values1, color='skyblue', label='Subgroup 1', alpha=0.7)
plt.bar(categories, values2, bottom=values1, color='orange', label='Subgroup 2', alpha=0.7)
plt.bar(categories, values3, bottom=np.array(values1) + np.array(values2), color='lightgreen', label='Subgroup 3', alpha=0.7)

# Add title and labels
plt.title('Stacked Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create an lmplot
sns.lmplot(x="total_bill", y="tip", data=data, hue="sex", markers=["o", "x"], palette="Set1")

# Add title
plt.title('Linear Regression Plot of Tips vs Total Bill')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("penguins")

# Create a KDE plot
sns.kdeplot(x=data['bill_length_mm'], y=data['bill_depth_mm'], fill=True, cmap="magma")

# Add title and labels
plt.title('2D KDE Plot of Bill Length and Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset('tips')

# Create a swarm plot
sns.swarmplot(x='day', y='total_bill', hue='sex', data=data, palette='Set2')

# Add title and labels
plt.title('Swarm Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,0])
y = np.array([4,5])
plt.quiver(x,y,angles='xy',scale_units='xy',scale=1)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title('vector visualisation')
plt.show()
```

```python
vector = np.array([1,5])
origin = np.array([2,5])

plt.quiver(vector,origin,angles='xy',scale_units='xy',scale=1)
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(-2, 2, 0.5), np.arange(-2, 2, 0.5))
U = -Y
V = X
plt.quiver(X, Y, U, V)
plt.show()
```

```python
origin = np.array([1,0])
vector = np.array([2,5])

plt.quiver(*origin,*vector,angles='xy',scale_units='xy',scale=1)

plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.title('visualising vectors')
plt.show()
```
