```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Vectors a and b
a = np.array([1, 0, 0])
b = np.array([0, 1, 0])

# Compute cross product
cross_product = np.cross(a, b)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector a
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='b', label='Vector a')

# Plot vector b
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='r', label='Vector b')

# Plot cross product vector
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='g', label='Cross Product')

# Set plot limits and labels
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X ax')

# Set the remaining axis labels
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the legend
ax.legend()

# Display the plot
plt.show()
```

# symmatric matrices

```python
b = np.array([[1,2,3],[2,5,6],[3,6,9]])
is_symmatric = np.array_equal(b,b.T)
print(is_symmatric)
```

```python
# Define a symmetric matrix
B = np.array([[1, 2, 3], [2, 5, 6], [3, 6, 9]])

# Check if B is symmetric
is_symmetric = np.array_equal(B, B.T)

print("\nMatrix B:")
print(B)
print(f"\nIs Matrix B symmetric? {is_symmetric}")
```

# skew symmatric 
if the negative of transpose is equal to matrix

```python
c = np.array([[0,-2,-3],[2,0,-6],[3,6,0]])
isskew = np.array_equal(c,-c.T)
print(c)
print(isskew)
```

```python
# Define a skew-symmetric matrix
C = np.array([[0, -2, -3], [2, 0, -6], [3, 6, 0]])

# Check if C is skew-symmetric
is_skew_symmetric = np.array_equal(C, -C.T)

print("\nMatrix C:")
print(C)
print(f"\nIs Matrix C skew-symmetric? {is_skew_symmetric}")
```

```python
a = np.array([[1,2],[4,5]])
a
inverse_a = np.linalg.inv(a)
inverse_a
```

```python
import numpy as np

# Define a 2x2 matrix
A_2x2 = np.array([[4, 3], [6, 2]])

# Calculate the determinant
det_A_2x2 = np.linalg.det(A_2x2)

print("Matrix A (2x2):")
print(A_2x2)
print(f"\nDeterminant of Matrix A (2x2): {det_A_2x2}")
```

```python
# Define a 3x3 matrix
B_3x3 = np.array([[1, 2, 3], [0, 1, 4], [5, 6, 0]])

# Calculate the determinant
det_B_3x3 = np.linalg.det(B_3x3)

print("\nMatrix B (3x3):")
print(B_3x3)
print(f"\nDeterminant of Matrix B (3x3): {det_B_3x3}")
```

```python
# Check if the determinant is non-zero
if det_A_2x2 != 0:
    # Calculate the inverse of the 2x2 matrix
    inverse_A_2x2 = np.linalg.inv(A_2x2)
    print("\nInverse of Matrix A (2x2):")
    print(inverse_A_2x2)
else:
    print("\nMatrix A (2x2) is singular and does not have an inverse.")

# Check if the determinant is non-zero
if det_B_3x3 != 0:
    # Calculate the inverse of the 3x3 matrix
    inverse_B_3x3 = np.linalg.inv(B_3x3)
    print("\nInverse of Matrix B (3x3):")
    print(inverse_B_3x3)
else:
    print("\nMatrix B (3x3) is singular and does not have an inverse.")
```
